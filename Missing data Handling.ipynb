def imputation(df,p,st):
    missing=pd.DataFrame(df.isnull().mean())
    #df1 =pd.DataFrame()
    for var in df.columns:       
        a=int(missing.loc[var])
        print(var,a)
        if(a==0):
            df = df
        elif(a>p):
            df1= df.drop([var])             
        elif(a<p):
            df1[var]=df1.groupby(st).transform(lambda x: x.fillna(x.mean()))             
    
    return df1

print(imputation(data,0.3,'Taluka'))


```python
missing=dict(data.isnull().mean())
```

```python
for key, value in missing.items():
    print(key, value)
    
    
```

```python
#serched variable which we want to drop >p = .3
lisim=[]
lisdr=[]
for key, value in missing.items():
 
    if(value<=0.3):
        lisim.append(key)
    else:
        lisdr.append(key)
        
print(lisim)
print(lisdr)
```

```python
df1=data.drop(lisdr, axis=1)
df=df1.groupby('Village').transform(lambda x: x.fillna(x.mean())) 
df[['TalukaName', 'Village', 'Irrigation', 'Revenue.Circle']] = df1[['TalukaName', 'Village', 'Irrigation', 'Revenue.Circle']]
df.info()
```

```python
df.to_csv ('imputed_cotton.csv', index = False, header=True)
```
